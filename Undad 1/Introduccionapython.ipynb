```python
import random
from statistics import mean, median, mode, stdev
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fábrica 1
fabrica1 = []
for i in range(100):
    numero_aleatorio = round(random.uniform(50, 100), 2)
    fabrica1.append(numero_aleatorio)

# Fábrica 2
fabrica2 = []
for i in range(100):
    numero_aleatorio2 = round(random.uniform(40, 93), 2)
    fabrica2.append(numero_aleatorio2)
```

```python
# Fábrica 1
maximo1 = max(fabrica1)
minimo1 = min(fabrica1)
rango1 = maximo1 - minimo1
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("Estadística descriptiva de la Fábrica 1")
print("Media =", media1)
print("Mediana =", mediana1)
print("Moda =", moda1)
print("Máximo =", maximo1)
print("Mínimo =", minimo1)
print("Rango =", rango1, "\n")

# Fábrica 2
maximo2 = max(fabrica2)
minimo2 = min(fabrica2)
rango2 = maximo2 - minimo2
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print("Estadística descriptiva de la Fábrica 2")
print("Media =", media2)
print("Mediana =", mediana2)
print("Moda =", moda2)
print("Máximo =", maximo2)
print("Mínimo =", minimo2)
print("Rango =", rango2, "\n")
```

```python
desviacion_estandar_fabrica1 = stdev(fabrica1)
print("Desviación estándar Fábrica 1:", desviacion_estandar_fabrica1)

desviacion_estandar_fabrica2 = stdev(fabrica2)
print("Desviación estándar Fábrica 2:", desviacion_estandar_fabrica2, "\n")
```

```python
plt.figure(figsize=(8,6))
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor="black")

plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label=f'Media={media1:.2f}')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana1:.2f}')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda1:.2f}')
plt.axvline(media1 - desviacion_estandar_fabrica1, color='black', linestyle='dashed',
            linewidth=2, label=f'Desv. Estándar ={desviacion_estandar_fabrica1:.2f}')
plt.axvline(media1 + desviacion_estandar_fabrica1, color='black', linestyle='dashed', linewidth=2)

plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la Fábrica 1")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(fabrica2, bins=10, alpha=0.2, color='b', edgecolor="black")

plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label=f'Media={media2:.2f}')
plt.axvline(mediana2, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana2:.2f}')
plt.axvline(moda2, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda2:.2f}')
plt.axvline(media2 - desviacion_estandar_fabrica2, color='black', linestyle='dashed',
            linewidth=2, label=f'Desv. Estándar ={desviacion_estandar_fabrica2:.2f}')
plt.axvline(media2 + desviacion_estandar_fabrica2, color='black', linestyle='dashed', linewidth=2)

plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la Fábrica 2")
plt.legend()
plt.show()
```

```python
sns.histplot(x=fabrica1, kde=True, kde_kws={'bw_adjust': 0.5}, bins=10)
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.title("Distribución de datos - Fábrica 1")
plt.show()
```
